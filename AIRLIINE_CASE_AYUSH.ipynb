# In This Type of Case use Statistical Model Instead of Using Traditional Machine Learning Model

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

**Additive Type**
**Multiplicative Type**

**It is Not AP,GP,HP**

```python
result=seasonal_decompose(air_1['Passengers'],model='add') # Additive Model
result.plot()
```

```python
result_1=seasonal_decompose(air_1['Passengers'],model='mul') # Multiplicative Model
result_1.plot()
```

**If Any Changes in Trend we found during mul & add negligible**

**After Performing ADDITIVE & MULTIPLICATIVE Methods - unable to find any changes as such**

**We have another two methods to selove the same**
**SMA- Simple Moving Average**
**EWMA-Exponential SMA**

```python
air_1['6-Months-SMA']=air_1['Passengers'].rolling(window=6).mean()
air_1['12-Months-SMA']=air_1['Passengers'].rolling(window=12).mean()
air_1['24-Months-SMA']=air_1['Passengers'].rolling(window=24).mean()
air_1['1080-Months-SMA']=air_1['Passengers'].rolling(window=1080).mean()
```

```python
air_1.plot()
```

**FROM SMA - WE FOUND DATA TREND BUT NEED TO DO ONE MORE TIME TESTING WITH EWMA TOO**

```python
air_1['12-Months-EWMA']=air_1['Passengers'].ewm(span=12,adjust=False).mean()
air_1['24-Months-EWMA']=air_1['Passengers'].ewm(span=24,adjust=False).mean()
air_1['1080-Months-EWMA']=air_1['Passengers'].ewm(span=1080,adjust=False).mean()
air_1['1200-Months-EWMA']=air_1['Passengers'].ewm(span=1200,adjust=False).mean()
```

```python
air_1.plot()
```
